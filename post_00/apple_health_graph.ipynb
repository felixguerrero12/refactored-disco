```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load the CSV into a DataFrame
# df = pd.read_csv('/content/drive/MyDrive/PublicBlogs/blog_001/health_data.csv')
df = pd.read_csv('health_data.csv')
```

```python
# Define the shorthand function
def extract_start_date(date_str):
    start, _ = date_str.split(' - ')  # Only take the start date and ignore the end date

    # Convert to datetime object
    start_date = datetime.strptime(start, '%Y-%m-%d %H:%M:%S')

    return start_date.strftime('%Y-%m-%d')
```

```python
# Apply the function to the "Date" column
df['Date'] = df['Date'].apply(extract_start_date)
```

```python
plt.figure(figsize=(15,7))
plt.bar(df['Date'], df['Active energy burned(Cal)'])
plt.xlabel('Date')
plt.ylabel('Active energy burned (Cal)')
plt.title('Active energy burned over Dates')
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.tight_layout()
plt.show()
```
